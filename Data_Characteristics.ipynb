### Abstract

Data on influenza-like illness (ILI) from 2010 to 2020 week 20 are included. The name of each column represents the year and the index represents the week. It also contains web data from the 35th week of 17 years to the 20th week of 20 years. For the United States and Australia, the dataset consists of influenza-related words in English and the rate of searches for those words from Google Trends. In the case of Korea, it consists of words related to influenza in Korean, the number of news and tweets containing the word, and the search rate of the word  from Naver Trends.
